```python
import pandas as pd
df = pd.read_excel('SerieA_2020-21.xlsx')
```

```python
df.head()
```

```python
# raggruppa per squadra e media i voti
df.groupby('Team').mean()
```

```python
# raggruppa per squadra e ruolo e media i voti
df.groupby(['Ruolo', 'Team']).mean()
```

```python
# raggruppa per squadra e ruolo e scegli il voto più alto
df.groupby(['Ruolo', 'Team']).max()
```
